```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sn
import ppscore as pps

from scipy.stats import zscore
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelBinarizer
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_auc_score, auc, classification_report, roc_curve

sn.set_style('darkgrid')
```

### Preprocessing  

```python
df = pd.read_csv('D:\\Phyton\\Weather_Data.csv')

X = df.select_dtypes(exclude='object')
y = df['RainToday']

def metadata(d):
    print('Review data')
    print(d.head(),'\n')
    print('Explanation of dataset')
    print(d.info(),'\n')
    print('Statistic of dataset')
    print(d.describe().T,'\n')
    print('Total missing data')
    print(d.isnull().sum(),'\n')
    print('Total duplicated data')
    print(d.duplicated().sum(),'\n')
    print('Skewness')
    print(d.skew(),'\n')
    print('Z score')
    print(X.apply(zscore),'\n')
    
metadata(df)
```

```python
x = StandardScaler().fit_transform(X)
y = LabelBinarizer().fit_transform(y).ravel()

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.3, random_state=0)
```

### Build model 

```python
model_rf = RandomForestClassifier()
model_sv = SVC(probability=True)
model_kn = KNeighborsClassifier()
model_nb = GaussianNB()

model_list = [model_rf, model_sv, model_kn, model_nb]
y_new = []
y_pel = []

for model in model_list:
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    y_new.append(y_pred)
    y_prob = model.predict_proba(x_test)
    y_pel.append(y_prob[:,1])
```
